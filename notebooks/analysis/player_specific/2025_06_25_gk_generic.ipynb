## Getting data

```python
df_shots = pd.read_csv(path_shots)
df_shots.head(2)
```

```python
df_player_matches = pd.read_csv(path_player_matches)
df_player_matches.head(2)
```

## Cleaning data

```python
df_shots = df_shots[~df_shots['player'].str.endswith('(pen)')]
df_shots['goal'] = (df_shots['outcome'] == 'Goal').astype('int')
df_shots['PSxG'] = df_shots['PSxG'].fillna(0)
df_shots.head(1)
```

## Filtering games

```python
if player_name is not None:
  df_player_matches = df_player_matches[df_player_matches['player'] == player_name]
  print(len(df_player_matches))
if season is not None:
  df_player_matches = df_player_matches[df_player_matches['season'].astype('str').isin(season)]
  print(len(df_player_matches))
df_player_matches = df_player_matches[df_player_matches['min'] >= 90]
print(len(df_player_matches))
len(df_player_matches)
```

```python
df_player_matches.head(1)
```

## Filtering shots faced

```python
df_player_matches = df_player_matches[['game', 'team']]
df_player_matches.columns = ['game', 'team_player']

df_player_shots = df_shots.merge(
  df_player_matches, on=['game'], how='inner'
)
df_player_shots = df_player_shots[df_player_shots['team'] != df_player_shots['team_player']]
print(len(df_player_shots))
df_player_shots.head(1)
```

## Grouping data

```python
print(player_name)

PSxG = sum(df_player_shots['PSxG'])
goals = sum(df_player_shots['goal'])
ratio = -1 * ((goals / PSxG) - 1)

print(f'PSxG: {PSxG}')
print(f'goals: {goals}')
print(f'% goals saved: {ratio:.2%}')
```

## Look at performance per difficulty

```python
df_player_shots_faced = df_player_shots[df_player_shots['outcome'].isin(['Saved', 'Goal'])]
print(len(df_player_shots_faced))
df_player_shots_faced['PSxG_cat'] = pd.cut(df_player_shots_faced['PSxG'], [0, 0.1, 0.2, 0.5, 1], labels=['0.00-0.10', '0.10-0.20', '0.20-0.50', '0.50-1.00'])
```

```python
df_player_shots_faced['count'] = 0
df_grouped = df_player_shots_faced.groupby('PSxG_cat',observed=True).agg({'count': 'count', 'goal': 'sum', 'PSxG': 'sum'})
df_grouped['% goals saved'] = -1 * ((df_grouped['goal'] / df_grouped['PSxG']) - 1)
df_grouped['% goals saved']

print(player_name)
df_grouped
```
